```python
import pandas as pd

data = pd.read_csv("insurance.csv")

data.head()
```

```python
print(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_columns = ['age', 'bmi', 'children', 'charges']
for col in numerical_columns:
    plt.figure()
    sns.histplot(data[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
from scipy.stats import mode

bmi_mean = data['bmi'].mean()
bmi_std = data['bmi'].std()
bmi_mode = mode(data['bmi']).mode[0]
charges_mean = data['charges'].mean()
charges_std = data['charges'].std()
charges_mode = mode(data['charges']).mode[0]

print(f'BMI Mean: {bmi_mean:.2f}, BMI Standard Deviation: {bmi_std:.2f}, BMI Mode: {bmi_mode:.2f}')

plt.figure()
sns.histplot(data['bmi'], kde=True)
plt.axvline(bmi_mean, color='r', linestyle='dashed', label=f'Mean: {bmi_mean:.2f}')
plt.axvline(bmi_mode, color='g', linestyle='dashed', label=f'Mode: {bmi_mode:.2f}')
plt.axvline(bmi_mean + bmi_std, color='b', linestyle='dashed', label=f'Std Dev: {bmi_std:.2f}')
plt.title('BMI Histogram with Mean, Mode, and Standard Deviation')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.legend()
plt.show()


print(f'Charges Mean: {charges_mean:.2f}, Charges Standard Deviation: {charges_std:.2f}, Charges Mode: {charges_mode:.2f}')

plt.figure()
sns.histplot(data['charges'], kde=True)
plt.axvline(charges_mean, color='r', linestyle='dashed', label=f'Mean: {charges_mean:.2f}')
plt.axvline(charges_mode, color='g', linestyle='dashed', label=f'Mode: {charges_mode:.2f}')
plt.axvline(charges_mean + charges_std, color='b', linestyle='dashed', label=f'Std Dev: {charges_std:.2f}')
plt.title('Charges Histogram with Mean, Mode, and Standard Deviation')
plt.xlabel('Charges')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Box plots for numerical indicators
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[numerical_columns])
plt.title('Box Plot for Numerical Indicators')
plt.xlabel('Feature')
plt.ylabel('Value')
plt.show()
```

```python
import numpy as np

n_samples = 300
sample_lengths = [10, 30, 100]

for n in sample_lengths:
    sample_means = []
    for _ in range(n_samples):
        sample = np.random.choice(data['charges'], size=n)
        sample_means.append(sample.mean())

    # Plot histograms of sample means
    plt.figure()
    sns.histplot(sample_means, kde=True)
    plt.title(f'Sample Means (n={n})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')
    plt.show()

    # Calculate mean and standard deviation of sample means
    sample_mean_mean = np.mean(sample_means)
    sample_mean_std = np.std(sample_means, ddof=1)

    print(f'Sample Mean (n={n}) Mean: {sample_mean_mean:.2f}, Std Dev: {sample_mean_std:.2f}')
```

```python
from scipy import stats

confidence_levels = [0.95, 0.99]

for level in confidence_levels:
    confidence_interval_bmi = stats.norm.interval(level, loc=bmi_mean, scale=bmi_std/len(data))
    confidence_interval_charges = stats.norm.interval(level, loc=charges_mean, scale=charges_std/len(data))

    print(f'{level * 100}% Confidence Interval for BMI: {confidence_interval_bmi}')
    print(f'{level * 100}% Confidence Interval for Charges: {confidence_interval_charges}')
```

```python
from scipy.stats import kstest, norm, probplot

def check_normality(data, col_name):
    ks_statistic, ks_p_value = kstest(data, norm(loc=data.mean(), scale=data.std()).cdf)
    probplot(data, plot=plt)
    plt.title(f'Q-Q Plot for {col_name}')
    plt.show()
    print(f'KS-Test for {col_name}: KS Statistic = {ks_statistic}, p-value = {ks_p_value}')

check_normality(data['bmi'], 'BMI')
check_normality(data['charges'], 'Charges')
```

```python
data = pd.read_csv("ECDCCases.csv")

# Check for missing values and calculate the percentage
missing_percentage = (data.isnull().sum() / len(data)) * 100

# Identify the two attributes with the most missing values
missing_columns = missing_percentage.nlargest(2).index.tolist()

# Drop the columns with the most missing values
data = data.drop(columns=missing_columns)

# Fill missing values for remaining columns
categorical_features = data.select_dtypes(include=['object']).columns
numeric_features = data.select_dtypes(exclude=['object']).columns

# Fill missing values for categorical features with 'Other'
data[categorical_features] = data[categorical_features].fillna('Other')

# Fill missing values for numeric features with the median
data[numeric_features] = data[numeric_features].fillna(data[numeric_features].median())
```

```python
# Display statistics of the dataset
statistics = data.describe()

# Identify columns with potential outliers
outlier_columns = ['cases', 'deaths']

# Identify days when the number of deaths exceeded 3000 for each country
high_death_days = data[data['deaths'] > 3000][['dateRep', 'countriesAndTerritories']]

# Count the number of such days for each country
high_death_days_count = high_death_days.groupby('countriesAndTerritories').size()

print(statistics)
print(f"Days with deaths > 3000:\n{high_death_days_count}")
```

```python
data = pd.read_csv("bmi.csv")

northwest = data[data.region == 'northwest']
southwest = data[data.region == 'southwest']
n = northwest.bmi
s = southwest.bmi
print("ShapiroWilk test")
test1 = stats.shapiro(n)
test2 = stats.shapiro(s)
print(f'northwest:\n {test1}\n southwest:\n {test2}')
print("\nBartlett's test")
print(stats.bartlett(s,n))
print("\nStudent's t-test")
print(stats.ttest_ind(s,n))
```
